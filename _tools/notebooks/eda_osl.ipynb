## Масштаб и пропуски
Смешение валют (USD Полюс / RUB сталевары) и разреженные объёмы — ключевые особенности данных.

```python
df.groupby('issuer').agg(
    n=('period', 'size'),
    currency=('currency', 'first'),
    target_min=('target', 'min'),
    target_max=('target', 'max'),
)
```

```python
# доля пропусков по колонкам-драйверам
cols = ['vol_copper_t','vol_nickel_t','vol_pd_oz','vol_pt_oz','vol_gold_oz','vol_steel_t',
        'p_gold','p_copper','p_nickel','p_platinum','p_steel_proxy_iron_ore','p_usd_rub']
(df[cols].isna().mean() * 100).round(0).rename('% NaN')
```
